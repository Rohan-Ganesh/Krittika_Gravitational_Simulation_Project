```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python


def Object_Dropped():
  v0 = 0.0
  g = 9.8
  h = 1e-4

  Hin = input('Enter the Height: ')
  H = None
  try:
    H = float(Hin)
  except:
    print('Incorrect Input')
    return
  if H < 0:
    print('Incorrect Input')
    return

  varr = [v0]
  yarr = [H]
  y = H
  v = v0

  while(y > 0):
    y += v*h
    v -= g*h
    yarr.append(y)
    varr.append(v)
  plt.plot(h*np.arange(0, len(yarr)), yarr)
  plt.xlabel('Time')
  plt.ylabel('Height')
  plt.title('Height vs Time for dropped object')
Object_Dropped()
```

```python
def Earth_Around_Sun():
  Me = 5.9722e24
  G = 6.67e-11
  Ms = 1.9885e30
  h = (1e-2)*(3.154e7)
  #Assuming reduced mass is same as mass of earth due to very low Me/Ms ratio
  C = G*Ms
  Rin = input('Enter the Radius in AU : ')
  R = None
  try:
    R = float(Rin)*1.496e11
  except:
    print('Incorrect Input')
    return
  if R < 0:
    print('Incorrect Input')
    return
  v0t = (C/R)**0.5
  v0R = 0.0
  phi0 = 0.0
  rarr = [R]
  phiarr = [phi0]
  vt = v0t
  vr = v0R
  # tarr =np.arange(0, 100, h)
  T = 100.0*(3.154e7)
  r = R
  phi = phi0
  t = 0.0

  while(t < T):
    phi += h*vt/r
    rp = rarr[-1]
    r += h*vr
    vr -= h*(C/(rp**2) - vt*vt/rp)
    vt -= h*(vt*vr/rp)
    rarr.append(r)
    phiarr.append(phi)
    t += h

  fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
  ax.plot(phiarr, np.array(rarr)/1.496e11)
  ax.set_title("Orbit of the earth by Euler Method")
  plt.show()

  v0t = (C/R)**0.5
  v0R = 0.0
  phi0 = 0.0
  rarr = [R]
  phiarr = [phi0]
  vt = v0t
  vr = v0R
  # tarr =np.arange(0, 100, h)
  T = 100.0*(3.154e7)
  r = R
  phi = phi0
  t = 0.0
  while(t < T):
    k1r = h*vr
    k1phi = h*vt/r

    l1r = h*(-1*C/(r**2) + vt*vt/r)
    l1phi = h*(-1*vt*vr/r)

    k2r = h*(vr + l1r/2)
    k2phi = h*((vt + l1phi/2)/r)

    l2r = h*(-1*C/(r + k1r/2)**2 + (vt + l1phi/2)*(vt + l1phi/2)/(r + k1r/2))
    l2phi = h*(-1*(vt + l1phi/2)*(vr + l1r/2)/(r + k1r/2))

    k3r = h*(vr + l2r/2)
    k3phi = h*((vt + l2phi/2)/(r + k2r/2))

    l3r = h*(-1*C/(r + k2r/2)**2 + (vt + l2phi/2)*(vt + l2phi/2)/(r + k2r/2))
    l3phi = h*(-1*(vt + l2phi/2)*(vr + l2r/2)/(r + k2r/2))

    k4r = h*(vr + l3r)
    k4phi = h*((vt + l3phi)/(r + k3r))

    l4r = h*(-1*C/(r + k3r)**2 + (vt + l3phi)*(vt + l3phi)/(r + k3r))
    l4phi = h*(-1*(vt + l3phi)*(vr + l3r)/(r + k3r))

    r += (1.0/6.0)*(k1r + 2*k2r + 2*k3r + k4r)
    vr += (1.0/6.0)*(l1r + 2*l2r + 2*l3r + l4r)
    phi += (1.0/6.0)*(k1phi + 2*k2phi + 2*k3phi + k4phi)
    vt += (1.0/6.0)*(l1phi + 2*l2phi + 2*l3phi + l4phi)
    rarr.append(r)
    phiarr.append(phi)
    
    t += h

  fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
  ax.plot(phiarr, np.array(rarr)/1.496e11)
  ax.set_title("Orbit of the earth by RK4 Method")
  plt.show()

Earth_Around_Sun()






```
